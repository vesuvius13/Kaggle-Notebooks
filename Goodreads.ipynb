```python
import pandas as pd
import numpy as np
import glob
pd.options.mode.chained_assignment = None
```

```python
import os
list_of_files = os.listdir(os.getcwd()) #list of files in the current directory
all_data =[]
for each_file in list_of_files:
    if each_file.startswith('book'):  #since its all type str you can simply use startswith
        print (each_file)
        df = pd.read_csv(each_file, usecols = ['Name', 'Rating', 'PublishYear', 'Authors'])
        all_data.append(df)
```

```python
data = pd.concat(all_data, axis = 0)
```

```python
new_data['Rating_sofis'] = np.where((new_data['Rating_y'] == 'it was amazing') | (new_data['Rating_y'] == 'really liked it'),
                                   float(4.5), np.where(new_data['Rating_y'] == 'liked it', float(3.8), 
                                                       np.where(new_data['Rating_y'] == 'it was ok', float(3.5), float(2.0))))
```

```python
new_data['Rating_mean'] = ((new_data['Rating_x'] + new_data['Rating_sofis'])/2).round(2).astype(float)
```
